```python
from osgeo import gdal
from matplotlib import pyplot as plt
import numpy as np
import h5py
```

```python
dataset = gdal.Open("./S1A_IW_SLC__1SDV_20210703T002756_20210703T002823_038606_048E2D_BE12.SAFE/measurement/s1a-iw2-slc-vv-20210703t002757-20210703t002822-038606-048e2d-005.tiff", gdal.GA_ReadOnly)
data=dataset.ReadAsArray(xoff=1900,yoff=1600,xsize=500, ysize=400)
data=abs(data)
data[data==0]=np.nan
```

```python
#plt.figure(figsize = (20,20))
plt.imshow(20*np.log10(data), cmap='gray',interpolation=None, origin='lower')
plt.colorbar()
```
